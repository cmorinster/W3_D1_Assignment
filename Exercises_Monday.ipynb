```python
### Exercise 1 - Turn the shopping cart program into an object-oriented program
```

```python
# Create a class called cart that retains items and has methods to add, remove, and show

class Cart():
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.cart = []
        
    def add_to_cart(self, item):
        self.cart.append(item)
    
    def remove_from_cart(self):
        new_bool = True
        while new_bool == True:
            item_r = input("what item would you like to remove?\n").upper()
            for item3 in self.cart:
                if item_r == item3.item:
                    self.cart.remove(item3)
                    print(f"{item_r.capitalize()} removed\n")
                    new_bool = False
                    
            else:
                print(f"{item_r.capitalize()} not found on shopping list\n")
                
    
    def show_cart(self):
        print("Cafe Chris Order:\n\n")
        for item1 in self.cart:
            if len(item1.item)>7:
                print(f"{item1.item.capitalize()}\t\t\t${item1.price}\n")
            else:
                print(f"{item1.item.capitalize()}\t\t\t\t${item1.price}\n")
    
    
class Item():
    
    def __init__(self, item):
        self.item = item
        self.price = input(f"What is the price of the {self.item.capitalize()}?\n").upper()
    
def run():
    print("Welcome to Cafe Chris!\n")
    cart1 = Cart(input("Can I have a name for the order?\n"))
    total = 0.00
    prompt = ''
    
    while prompt.upper() != "DONE":
        
        prompt = input("What would you like to do?\nAdd Items?  (Type: Add)\nDelete Items?  (Type: Delete)\nSee Current List?  (Type: List)\nCheck Out?  (Type: Done)\n")
        if prompt.upper() == "ADD":
            new_item = Item(input("What item would you like to add?\n").upper())
            cart1.add_to_cart(new_item)
            
        elif prompt.upper() == "DELETE":
            cart1.remove_from_cart()
           
        elif prompt.upper() == "LIST":
            cart1.show_cart()
                
                
    print("Cafe Chris Order:\n\n")
    print("Item:\t\t\t\tPrice:")
    for item2 in cart1.cart:
        if len(item2.item)>7:
            print(f"{item2.item.capitalize()}\t\t\t${item2.price}\n")
        else:
            print(f"{item2.item.capitalize()}\t\t\t\t${item2.price}\n")
        total = total + float(item2.price)
    tax = total*0.05
    total = tax + total
    print(f"Tax:\t\t\t\t${tax:.2f}")
    print(f"Total:\t\t\t\t${total:.2f}\n")
    print(f"Thank You {cart1.customer_name} for your Order")
    print("We will call your name when the order is ready!")
          
          
    
        
        
        
        
                    

run()
```

```python
###Exercise 2 - Write a Python class for an Animal that has a name and energy attributes. The animal class should also have methods for eat, sleep, and play that will take in an integer and increase/decrease the energy of the animal with a formatted print statement
```

```python
class Animal():
    def __init__(self, name, energy=50):
        self.name = name
        self.energy = energy
        print(f"Hello {self.name} welcome to the Animal Daycare Center")
    def eat(self, food_total):
        self.energy = int(self.energy + food_total*0.4)
        print(f"By eating {food_total} ounces of food. {self.name}'s energy has increased to {self.energy}")
    def play(self, mins):
        self.energy = int(self.energy - mins*0.3)
        print(f"That sure was fun, but playing for {mins} minutes lowered {self.name}'s energy to {self.energy}.")
        if self.energy <= 0:
            print(f"{self.name} needs to eat and rest!!")
    def sleep(self, mins):
        self.energy = int(self.energy + mins*0.5)
        print(f"A good {mins} minute nap has increased {self.name}'s energy to {self.energy}!")
        
dog1 = Animal("Baxter")
dog2 = Animal("Felix", 30)
dog3 = Animal("Rex", 20)

dogs = [dog1, dog2, dog3]
for dog in dogs:
    dog.play(80)
    dog.eat(22)
    dog.sleep(60)

  
              
        
```
